```python
#reading elements in video
import pathlib
import platform
import torch
import cv2
import ssl

# 1. Fix the Linux-to-Windows path issue
if platform.system() == 'Windows':
    pathlib.PosixPath = pathlib.WindowsPath

# 2. Fix the SSL connection issue
ssl._create_default_https_context = ssl._create_unverified_context

# 3. Load the model
model = torch.hub.load('ultralytics/yolov5', 'custom', 
                       path='C:/Users/AMIT/Desktop/cvvvv/best.pt', 
                       force_reload=False)

# Optional: Use GPU if available for a major speed boost
if torch.cuda.is_available():
    model.cuda()

# 4. Open the video
video_path = 'C:/Users/AMIT/Desktop/cvvvv/2026-04-05 14-58-44.mp4'
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # --- SPEED OPTIMIZATION: Resize ---
    # Resizing to 640x480 is the native resolution for most YOLOv5 models.
    # This prevents the CPU from having to process high-resolution frames (1080p/4k).
    frame = cv2.resize(frame, (640, 480))

    # Inference
    results = model(frame)

    # Render results on the frame
    results.render() 
    
    # Show window
    cv2.imshow("YOLO Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
