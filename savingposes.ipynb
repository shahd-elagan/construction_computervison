```python
# this code is to cut videos into frames 
import cv2
import torch
import os
import pathlib
import platform

# 1. Windows Path Fix
if platform.system() == 'Windows':
    pathlib.PosixPath = pathlib.WindowsPath

# 2. Load YOLO Model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='C:/Users/AMIT/Desktop/cvvvv/best.pt')

# 3. Configuration
video_path = 'C:/Users/AMIT/Desktop/cvvvv/video5.mp4'
output_root = 'C:/Users/AMIT/Desktop/cvvvv/Dataset_Unlabeled2'
TARGET_CLASS = 2         # Only capture Class 2 (Excavator)
CLIP_LENGTH = 300       # 300 frames = 10 seconds (at 30 FPS)
SKIP_BETWEEN_CLIPS = 150 # Wait 5 seconds between clips to get different actions
CONF_THRESHOLD = 0.5

if not os.path.exists(output_root):
    os.makedirs(output_root)

cap = cv2.VideoCapture(video_path)
clip_count = 0
frame_idx = 0

print(f"--- Processing Video for Class {TARGET_CLASS} (10-second clips) ---")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    
    # Trigger a new 10-second recording session
    if frame_idx % (SKIP_BETWEEN_CLIPS + CLIP_LENGTH) == 0:
        clip_count += 1
        current_clip_folder = os.path.join(output_root, f"clip_{clip_count:03d}")
        
        print(f"Checking for Excavator to start Clip {clip_count}...")

        # Inner loop to capture 300 frames in a row
        for i in range(CLIP_LENGTH):
            if not ret: break
            
            # 4. Run AI Inference
            results = model(frame)
            df = results.pandas().xyxy[0]
            
            # --- NEW: FILTER FOR CLASS 2 AND CONFIDENCE ---
            # 'class' is the column name for the ID in YOLOv5 pandas output
            df_excavator = df[(df['class'] == TARGET_CLASS) & (df['confidence'] >= CONF_THRESHOLD)]

            if not df_excavator.empty:
                # Ensure folder is only created if we actually see the excavator
                if not os.path.exists(current_clip_folder):
                    os.makedirs(current_clip_folder, exist_ok=True)

                row = df_excavator.iloc[0] # Take the highest confidence excavator
                x1, y1, x2, y2 = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax'])
                
                # 5. Crop and Resize
                crop = frame[y1:y2, x1:x2]
                if crop.size > 0:
                    # 224x224 is the standard size for most Activity Recognition models
                    crop_resized = cv2.resize(crop, (224, 224))
                    cv2.imwrite(os.path.join(current_clip_folder, f"frame_{i:03d}.jpg"), crop_resized)
            
            # Move to the next frame within the 10-second window
            ret, frame = cap.read()
            frame_idx += 1

    frame_idx += 1

cap.release()
print(f"Finished! Check folders in: {output_root}")
```
